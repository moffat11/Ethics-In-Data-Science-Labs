# 1. Pre-processing the Data

In lab 4, you learned about *linear regression* and how it's used to make predictions on data when the labels are continuous values.  For data where the outcomes are discrete, we use classification, and specifically here, *logistic regression*.

**Question 1.1.** In the cell below, import the dataset `iris.csv` and assign it to a DataFrame called `iris_df`. Print its first 5 values.

```python
# YOUR CODE HERE
iris_im = pd.read_csv("iris.csv")
iris_df = pd.DataFrame(iris_im)
iris_df.head()
```

```python
# TEST CELL: do not delete!
```

This dataset contains three iris species: setosa, versicolor, and virginica. It also contains the flower's petal length, petal width, sepal length, and sepal width. Our goal is to use logistic regression so that we can predict what species of iris we have given its measurements.

**Question 1.2.** Recall that we denote `X` as our features (the measurements) and `Y` as our target (the species).

Using pandas functions, separate the `iris_df` DataFrame into these two components:


1.   Assign all the feature columns to a new DataFrame called `X`.
2.   Assign the single target column to a new Series called `Y`.



```python
# YOUR CODE HERE
X = iris_df[["Id", "SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm" ]]
Y = pd.Series(iris_df["Species"])
```

```python
# TEST CELL: do not delete!
```

Luckily, we don't have to make our own logistic regression model from scratch. We can import a library called `sklearn` that has the functions we need to make predictions on our iris data. However, for the functions to work, our target data must be numerical. To convert categorical data to numerical, we encode the target labels with values between 0 and n-1, where n is the number of classes. For our iris data, we will use sklearn's [LabelEncoder](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html).

**Question 1.3.** First import sklearn's LabelEncoder, then convert the targets from categorical to numerical. Assign your answer to `Y`. Consult the link above for help.

```python
# YOUR CODE HERE
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(["Iris-setosa","Iris-versicolor","Iris-virginica"])

target_transform = le.transform(iris_df["Species"])
Y = pd.Series(target_transform, name = 'Species')
```

```python
# TEST CELL: do not delete!
```

# 2. Pipeline

A common and powerful workflow in machine learning is to chain multiple processing steps (like scaling data and then training a model) into a single object. This is what scikit-learn's `Pipeline` does.

A crucial step before we can properly train and evaluate any model (including a pipeline) is to split our data into two sets:

1. Training set: This subset (we'll use 80%) is used to fit the model. The model learns patterns from this data.
2. Testing set: This "held-out" subset (the remaining 20%) is used to evaluate how well the model performs on new, unseen data.

It's vital that the test data is never used for training. Using it would be like giving a student the answers to an exam before they take it. The model would "memorize" the test data (a problem called **overfitting**), and we wouldn't have a reliable way to know if it can generalize to solve new problems.

**Question 2.1.** First import sklearn's [train_test_split](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html). Next, use this function to split your `X` (features) and `Y` (target) data. The function will return four new datasets; assign them to `X_train`, `X_test`, `Y_train`, and `Y_test`, respectively. The parameter `test_size` sets the proportion of data for the testing subset (set to 0.2). The `random_state` parameter controls how much the data is shuffled before being split.

```python
# YOUR CODE HERE
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=11)
```

The [Pipeline](https://scikit-learn.org/stable/modules/generated/sklearn.pipeline.Pipeline.html) is an extremely useful sklearn object that lets you chain multiple data-processing steps into a single workflow.

For this lab, our pipeline will perform **two steps** in sequence:
1. **StandardScaler**: This is a data transformer that standardizes our features. It scales the data in each column to have a mean of 0 and a standard deviation of 1. This is a common and important preprocessing step, as many models (including logistic regression) perform better when all features are on a similar scale.

2. **Logistic Regression**: This is our estimator (the model). After the StandardScaler transforms the data, the newly scaled data will be passed directly to this model for training. [Logistic Regression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html)

**Question 2.2.** Import the appropriate functions for the pipeline, logistic regression, and standard scalar. Next, create the `Pipeline` object and assign it to the variable `pipeline`. The pipeline is defined by a list of `(name, object)` tuples for **each step**.



*  The first step should be named `'scaler'` and use an instance of `StandardScaler()`.
*   The second step should be named `'classifier'` and use an instance of `LogisticRegression()`. Remember to set the `max_iter` parameter to 200 inside `LogisticRegression()`

```python
# YOUR CODE HERE
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

pipeline = Pipeline([('scaler', StandardScaler()), ('classifier', LogisticRegression(max_iter=200))])
```

**Question 2.3.** Now, use the `.fit` method to fit the training data to our pipeline. The output will be a diagram that shows each step our pipeline takes to transform and fit the data to the logistic regression model. Click on the steps to expand them.

```python
# YOUR CODE HERE
pipeline.fit(X_train, Y_train)
```

# 3. Predicting and Assessing

Now that our pipeline has been **fitted** (trained) using the training data, it's ready to make predictions on new, unseen data.

**Question 3.1.** Use the pipeline's `.predict` method to make predictions and assign your answer to `Y_pred`. Make sure you use the right feature subset (training vs. testing)!

```python
# YOUR CODE HERE
Y_pred = pipeline.predict(X_test)
print(X_test)
print("\nY_pred: ", Y_pred)
```

We print the X_test subset and Y_pred to better understand what's going on. X_test contains 30 random flowers and their metrics. Y_pred is an array of 30 predictions where 0 denotes setosa, 1 denotes versicolor, and 2 denotes verginica.

Now let's assess the performance of our predictor. The first tool that we can use is the [confusion matrix](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html) to get the true positive, true negative, false positive, and false negative of our predictor.

**Question 3.2.** Import the confusion matrix function from `sklearn` and test the predictions made above. Assign your answer to `conf_matrix`.

```python
# YOUR CODE HERE
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(Y_test, Y_pred)
print("Confusion Matrix:\n", conf_matrix)
```

Now, run the cell below to better visualize the confusion matrix:

```python
# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

`sklearn` provides a function called [classification report](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.classification_report.html), which calculates the `precision`, `recall`, `F1-score`, and `support` of a predictor. The support is how many samples there are in each class.

**Question 3.3.** Import the classification report from `sklearn` and use it to assess the performance of your predictions. Assign your answer to `class_report`.

```python
# YOUR CODE HERE
from sklearn.metrics import classification_report
class_report = classification_report(Y_test, Y_pred)
print("\nClassification Report:\n", class_report)
```

**Question 3.4.** What is the precision, recall, and F1-score of each iris species? What does this tell us about the performance of the predictor?

For all the three species, the precision, recall and F1 score is at 1.0. The predictor was able to predict the correct species for all the cases in the test data; hence the predictor is accurate and highly precise.
